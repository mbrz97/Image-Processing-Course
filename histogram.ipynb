```python
# Import necessary libraries
import matplotlib.pyplot as plt
from google.colab import files
from PIL import Image
import numpy as np

# Function to upload an image
def upload_image():
    uploaded = files.upload()
    filename = next(iter(uploaded))
    image = Image.open(filename)
    return image

# Function to plot histogram of the image
def plot_histogram(image):
    # Convert image to numpy array
    image_array = np.asarray(image)

    # If the image has three channels (RGB), separate them
    if len(image_array.shape) == 3:
        colors = ('r', 'g', 'b')
        for i, color in enumerate(colors):
            # Calculate histogram for each channel
            hist, bins = np.histogram(image_array[:,:,i], bins=256, range=(1, 256))
            plt.plot(bins[:-1], hist, color=color)
        plt.title('Histogram for Red, Green, and Blue channels')
    else:
        # Calculate histogram for grayscale image
        hist, bins = np.histogram(image_array, bins=256, range=(0, 256))
        plt.plot(bins[:-1], hist, color='black')
        plt.title('Histogram for Grayscale Image')

    plt.xlabel('Pixel Intensity')
    plt.ylabel('Pixel Count')
    plt.show()

# Main function to run the process
def main():
    print("Please upload an image:")
    image = upload_image()
    plot_histogram(image)

# Run the main function
main()
```
