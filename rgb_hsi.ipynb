```python
import cv2
import numpy as np
from google.colab import files
import matplotlib.pyplot as plt

def upload_image():
    uploaded = files.upload()
    filename = next(iter(uploaded))
    image = cv2.imread(filename)
    return image

def convert_to_rgb(image):
    return cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

def split_color_channels_and_convert_to_grayscale(rgb_image):
    red_channel = rgb_image[:, :, 0]
    green_channel = rgb_image[:, :, 1]
    blue_channel = rgb_image[:, :, 2]

    return red_channel, green_channel, blue_channel

def convert_to_hsv_and_split(bgr_image):
    hsv_image = cv2.cvtColor(bgr_image, cv2.COLOR_BGR2HSV)
    h, s, v = cv2.split(hsv_image)
    return h, s, v

def show_images(images, titles):
    n = len(images)
    plt.figure(figsize=(20, 10))
    for i in range(n):
        plt.subplot(1, n, i + 1)
        plt.imshow(images[i], cmap='gray')
        plt.title(titles[i])
        plt.axis('off')
    plt.show()

image_bgr = upload_image()
image_rgb = convert_to_rgb(image_bgr)

red_channel, green_channel, blue_channel = split_color_channels_and_convert_to_grayscale(image_rgb)
h_channel, s_channel, v_channel = convert_to_hsv_and_split(image_bgr)

images = [image_rgb, red_channel, green_channel, blue_channel, h_channel, s_channel, v_channel]
titles = ["Original Image", "Red Channel (Grayscale)", "Green Channel (Grayscale)", "Blue Channel (Grayscale)", "Hue Channel", "Saturation Channel", "Value Channel"]

show_images(images, titles)
```
